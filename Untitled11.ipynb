```python
#Importing libraries

import pandas as pd 
import numpy as np 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing dataset and viewing first 5 entries

df = pd.read_csv("C:/Kaggle Data/melt/diabetic_data.csv")
df.head()
```

```python
#Sorting dataset and converting it into dataframe

sort= df.iloc[:, 12:18].values
dfn = pd.DataFrame(sort)
print(dfn)
```

```python
#Using KMeans to iterate over given dataset  

distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(dfn)
    distortions.append(kmeanModel.inertia_) #Calculating the least square errors for each value
```

```python
#Visualizing the relationship between the columns 

g = sns.PairGrid(data=dfn)
g.map(sns.scatterplot) #Mapping all the column data using scatterplot
```

```python
#Plotting the elbow method for predicting optimal K value

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

***

From the above elbow cure, we can determine the optimal number of cluster. We can select the value of 'k' point, after which the distortion starts decreasing in linear manner. 

Hence for the given data, we can conclude that the **optimal number of clusters for the data is 2**.

***

```python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(dfn) #We transform the sorted dataset to scale the data and lean the scaling parameter

kmeanModel = KMeans(n_clusters=2, max_iter = 100) #Clustering using KMeans method
kmeanModel.fit(X_std) #Fit is used to perform calculation on feature values of data, and fit this calculation to transformer
center = kmeanModel.cluster_centers_ #Calculating the coordinates of the cluster centers
center
```

```python
plt.scatter(X_std[:,0], X_std[:,2]) #Plotting the best data using scatter method

# Plot the clusters centers
plt.scatter(kmeanModel.cluster_centers_[:, 0], kmeanModel.cluster_centers_[:, 1], 
            s=200,                             # Set centroid size
            c='red')                           # Set centroid color
plt.show()
```

```python
sns.scatterplot(x=X_std[:,0], y=X_std[:,2], hue=kmeanModel.labels_) #Using scatterplot to plot the data
plt.scatter(kmeanModel.cluster_centers_[:,0], kmeanModel.cluster_centers_[:,1], marker="X", c="r", s=80, label="centroids") #Plotting the center of the data
plt.legend()
plt.show()
```

Using KMeans, we can plot the data for second iteration and here we can clearly see that data is seperated into two different clusters. Hence we can say that k=2 gives the best result.
